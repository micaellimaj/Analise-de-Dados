Análise exploratória e visulização de dados, vamos começar importando **matplotlib.pyplot e seaborn.**

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**Mapa de Calor**

```python
# Atributos que serão correlacionados
raw_data = ['Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
# obtendo a matriz de correlação
corrmat = df[raw_data].corr()
# configurando o tamanho do plot
fig = plt.figure(figsize = (12,9))
# mascarando a parte triangular superior já que a matriz é simétrica(repetitiva)
mask = np.triu(np.ones_like(corrmat, dtype = bool))
sns.heatmap(corrmat, vmax = .8, mask=mask, square = True,annot = True)
plt.show()
```

O mapa de calor mostra a correlação entre os dados da população, quanto maior o valor correlacionado, maior a chance de serem dois dados correlacionados, mas não necessariamente.

Explore uma ou mais colunas traçando um gráfico abaixo e adicione alguma explicação sobre a matriz de correlação.

```python
plt.figure(figsize=(20,10))
changelist = ['Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
# gerar uma matriz de correlação para todo o conjunto de dados
corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

*   Quanto maior o valor, maior a chance de relação.

## Fazendo e Respondendo Perguntas:

**Q1: Descubra a variação normal da economia dos 10 primeiros paises**

```python
changelist = ["Economy (GDP per Capita)","Country"]

# Ordenar o dataframe pela coluna 'Economy (GDP per Capita)' em ordem decrescente
df_sorted = df.sort_values(by=changelist[0], ascending=False)

# Selecionar os 10 primeiros países
top_10_countries = df_sorted['Country'].head(10)
top_10_economies = df_sorted[changelist[0]].head(10)

plt.figure(figsize=(12,12))
plt.bar(top_10_countries, top_10_economies)
plt.xlabel('Country')
plt.ylabel('Economy (GDP per Capita)')
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Economy')
plt.show()
```

**Q2: encontre os indices socioeconômicos mais baixos**

```python
# Obter as cinco primeiras regiões únicas
top_5_regions = df['Region'].unique()[:5]

plt.figure(figsize=(10,10))
sociolist = ['Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
for i in range(len(sociolist)):
  plt.subplot(4,3,i+1)
  plt.subplots_adjust( hspace=1,wspace=0.5)
  plt.title(sociolist[i])
  plt.plot(df[sociolist[i]])
  plt.xticks(rotation=90)
plt.suptitle("Comparaçao Socioeconômica Mundial")
```

Na visualização acima, podemos notar que ocorre bastante variância de valores, às vezes em alta e outras em baixa, podendo dificultar alguns insights, no entanto, observa-se que há uma baixa nos valores de pontuação de felicidade, economia, família e expectativa de vida com o passar dos países(eixo x), assim podemos concluir que esses valores sociais interferem bastante no estilo de vida da população, por estarmos fazendo uma análise em cima de dados que estão em um raking decrescente de grau de felicidade, isso explica esses valores.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15))
sociolist = ['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom',
             'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']

# Obter as cinco primeiras regiões únicas
top_5_regions = df['Region'].unique()[:5]

titles = ['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom',
          'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']

for i in range(len(sociolist)):
    plt.subplot(4, 3, i + 1)
    plt.subplots_adjust(hspace=1, wspace=0.5)
    for region in top_5_regions:
        region_data = df[df['Region'] == region]
        plt.plot(region_data[sociolist[i]], label=region)
    plt.xticks(rotation=90)
    plt.legend(bbox_to_anchor=(1.50, 1), loc='upper left')
    plt.title(titles[i])

plt.suptitle("Comparação Socioeconômica Por Continente")
plt.show()
```

Acima podemos vizualizar informações sobre os tipos socioeconômicos da população divididos por região. Analisando os dados, podemos tirar várias conclusões sobre eles, como a Middle East and Northern Africa apresentar o maior valor de genoridade no eixo x(150) e um dos menores valores no eixo x(150) em família e expectativa de vida, por exemplo, entre outros insights que podem ser formados com uma boa observação.

Abaixo estamos comparando a liberdade com resíduo de distopia( ideia ou descrição de um país, de uma sociedade ou de uma realidade imaginários em que tudo está organizado de uma forma opressiva, assustadora ou totalitária, por oposição à utopia). Analisando o gráfico abaixo, podemos ver que o grau de distopia maior faz com que o de liberdade diminua.
